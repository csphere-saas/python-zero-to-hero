# Regular Expressions
A regular expression or RegEx is a special text string that helps to find patterns in data. A RegEx can be used to check if some pattern exists in a different data type. To use RegEx in python first we should import the RegEx module which is called re.
# <font color="blue">Expressions régulières</font>
<font color="blue">Une expression régulière, ou RegEx, est une chaîne de texte spéciale qui permet de trouver des modèles dans les données. Une RegEx peut être utilisée pour vérifier l'existence d'un modèle dans un autre type de données. Pour utiliser RegEx en Python, nous devons d'abord importer le module RegEx appelé re.</font>

### The re Module
After importing the module we can use it to detect or find patterns.

### <font color="blue">Le module re
<font color="blue">Après avoir importé le module, nous pouvons l'utiliser pour détecter ou trouver des modèles.</font>

```python
import re
```

### Methods in re Module
To find a pattern we use different set of re character sets that allows to search for a match in a string.
<ul>
<li><b>re.match(): </b>searches only in the beginning of the first line of the string and returns matched objects if found, else returns None.</li>
<li><b>re.search: </b>Returns a match object if there is one anywhere in the string, including multiline strings.</li>
<li><b>re.findall: </b>Returns a list containing all matches</li>
<li><b>re.split: </b>Takes a string, splits it at the match points, returns a list</li>
<li><b>re.sub: </b>Replaces one or many matches within a string</li>
</ul>

### <font color="blue">Méthodes du module re
<font color="blue">Pour trouver un motif, nous utilisons différents jeux de caractères re permettant de rechercher une correspondance dans une chaîne.
<ul><font color="blue">
<li><b>re.match() : </b>Recherche uniquement au début de la première ligne de la chaîne et renvoie les objets correspondants s'ils sont trouvés ; sinon, renvoie None.</li>
<li><b>re.search : </b>Renvoie un objet correspondant s'il en existe un dans la chaîne, y compris les chaînes multi-lignes.</li>
<li><b>re.findall : </b>Renvoie une liste contenant toutes les correspondances.</li>
<li><b>re.split : </b>Prend une chaîne, la divise aux points de correspondance et renvoie une liste.</li>
<li><b>re.sub : </b>Remplace une ou plusieurs correspondances dans une chaîne.</li>
</ul>

### match

```python
# syntac
re.match(substring, string, re.I)
# substring is a string or a pattern, string is the text we look for a pattern , re.I is case ignore
# substring est une chaîne ou un modèle, string est le texte dans lequel nous recherchons un modèle, re.I est ignorer la casse
```

```python
import re

txt = 'I love to teach python and javaScript'
# It returns an object with span, and match 
#Il renvoie un objet avec span et match
match = re.match('I love to teach', txt, re.I)
print(match)  
```

```python
# We can get the starting and ending position of the match as tuple using span
# Nous pouvons obtenir la position de début et de fin du match sous forme de tuple en utilisant span
span = match.span()
print(span)    
```

```python
# Lets find the start and stop position from the span
# Trouvons la position de départ et d'arrêt à partir de span

start,end = span
print(start, end)  
substring = txt[start:end]
print(substring)     
  
```

As you can see from the example above, the pattern we are looking for (or the substring we are looking for) is I love to teach. The match function returns an object only if the text starts with the pattern.
<br><font color="blue">Comme vous pouvez le voir dans l'exemple ci-dessus, le motif recherché (ou la sous-chaîne recherchée) est « J'aime enseigner ». La fonction match ne renvoie un objet que si le texte commence par ce motif.

```python
import re

txt = 'I love to teach python and javaScript'
match = re.match('I like to teach', txt, re.I)
print(match)
```

### Search

```python
import re

txt = '''Python is the most beautiful language that a human being has ever created.
I recommend python for a first programming language.'''

# It returns an object with span and match
# Il renvoie un objet avec span et match
match = re.search('first', txt, re.I)
print(match) 
```

```python
 
# We can get the starting and ending position of the match as tuple using span
# Nous pouvons obtenir la position de début et de fin du match sous forme de tuple en utilisant span
match = match.span()
print(match)     
```

As you can see, search is much better than match because it can look for the pattern throughout the text. Search returns a match object with a first match that was found, otherwise it returns None. A much better re function is findall. This function checks for the pattern through the whole string and returns all the matches as a list.
<br><font color="blue">Comme vous pouvez le constater, la fonction search est bien plus performante que la fonction match, car elle permet de rechercher le motif dans tout le texte. Search renvoie un objet match avec la première correspondance trouvée ; sinon, elle renvoie None. Une fonction re bien plus performante est findall. Cette fonction recherche le motif dans toute la chaîne et renvoie toutes les correspondances sous forme de liste.

Searching for All Matches Using findall
findall() returns all the matches as a list

<br><font color="blue">Rechercher toutes les correspondances avec findall<br>
findall() renvoie toutes les correspondances sous forme de liste

```python
txt = '''Python is the most beautiful language that a human being has ever created.
I recommend python for a first programming language'''

# It return a list
nombre_matches = re.findall('language', txt, re.I)
print(nombre_matches)  
```

```python
txt = '''Python is the most beautiful language that a human being has ever created.
I recommend python for a first programming language'''

# It return a list
nombres_matches = re.findall('beautiful language', txt, re.I)
print(nombres_matches)  
```

```python
txt = '''Python is the most beautiful language that a human being has ever created.
I recommend python for a first programming language'''

# It return a list
nombres_matches = re.findall('language', txt, re.I)
print(nombres_matches) 
```

### Writing RegEx Patterns
To declare a string variable we use a single or double quote. To declare RegEx variable r''. The following pattern only identifies apple with lowercase, to make it case insensitive either we should rewrite our pattern or we should add a flag.
### <font color="blue">Écriture de motifs RegEx
<font color="blue">Pour déclarer une variable chaîne, nous utilisons des guillemets simples ou doubles. Pour déclarer une variable RegEx, nous utilisons r''. Le motif suivant identifie uniquement les pommes en minuscules. Pour le rendre insensible à la casse, nous devons soit le réécrire, soit ajouter un indicateur.

```python
import re

regex_pattern = r'Apple'
txt = 'Apple and banana are fruits. An old cliche says an apple a day a doctor way has been replaced by a banana a day keeps the doctor far far away. '
matches = re.findall(regex_pattern, txt)
print(matches) 
```

```python
# To make case insensitive adding flag '
# Pour rendre l'ajout de l'indicateur « insensible à la casse »
matches = re.findall(regex_pattern, txt, re.I)
print(matches)  
```

```python
txt = '''Python is the most beautiful language that a human being has ever created.
I recommend python for a first programming language'''

test_sossou= r'Python|python'
matches = re.findall(test_sossou, txt)
print(matches) 
```

```python
# or we can use a set of characters method
# this mean the first letter could be Apple or apple / cela signifie que la première lettre pourrait être Apple ou apple
regex_pattern = r'[Aa]pple'  
matches = re.findall(regex_pattern, txt)
print(matches) 
```

### Square Bracket
Let us use square bracket to include lower and upper case

### <font color="blue">Crochets
<font color="blue">Utilisons des crochets pour inclure les minuscules et les majuscules.

```python
regex_pattern = r'[Aa]pple' # this square bracket mean either A or a / ce crochet signifie soit A ou a
txt = 'Apple and banana are fruits. An old cliche says an apple a day a doctor way has been replaced by a banana a day keeps the doctor far far away.'
matches = re.findall(regex_pattern, txt)
print(matches)  
```

```python
# If we want to look for the banana, we write the pattern as follows:
# Si nous voulons chercher la banane, nous écrivons le modèle comme suit :
regex_pattern = r'[Aa]pple|[Bb]anana' # this square bracket means either A or a
txt = 'Apple and banana are fruits. An old cliche says an apple a day a doctor way has been replaced by a banana a day keeps the doctor far far away.'
matches = re.findall(regex_pattern, txt)
print(matches)  
#Using the square bracket and or operator , we manage to extract Apple, apple, Banana and banana.
#En utilisant l'opérateur crochet et ou , nous parvenons à extraire Pomme, pomme, Banane et banane.
```

### Escape character(\) in RegEx

```python
regex_pattern = r'\d'  # d is a special character which means digits
txt = 'This regular expression example was made on December 6,  2019 and revised on July 8, 2021'
digits = re.findall(regex_pattern, txt)
print("this is not what we want", digits)  
```

### One or more times(+)

```python
# d is a special character which means digits, + mean one or more times
# d est un caractère spécial qui signifie chiffres, + signifie une ou plusieurs fois
regex_pattern = r'\d+'  
txt = 'This regular expression example was made on December 6,  2019 and revised on July 8, 2021'
matches = re.findall(regex_pattern, txt)
print("Now, this is better!",matches)
```

### Period(.)

```python
regex_pattern = r'[a].'  # this square bracket means a and . means any character except new line 
#ce crochet signifie a et . signifie n'importe quel caractère sauf nouvelle ligne
txt = '''Apple and banana are fruits'''
matches = re.findall(regex_pattern, txt)

print(matches)  
```

```python
regex_pattern = r'[a].+'  
# . any character, + any character one or more times 
#. n'importe quel caractère, + n'importe quel caractère une ou plusieurs fois
matches = re.findall(regex_pattern, txt)
print(matches)  
```

### Zero or more times(*)
Zero or many times. The pattern could may not occur or it can occur many times.

### <font color="blue">Zéro ou plusieurs fois(*)
<font color="blue">Zéro ou plusieurs fois. Le motif peut ne pas se reproduire ou se reproduire plusieurs fois.

```python
regex_pattern = r'[a].*'  
# . any character, * any character zero or more times 
## . n'importe quel caractère, * n'importe quel caractère zéro ou plusieurs fois
txt = '''Apple and banana are fruits'''
matches = re.findall(regex_pattern, txt)
print(matches) 
```

```python
regex_pattern = r'[z].*'  
# . any character, * any character zero or more times 
## . n'importe quel caractère, * n'importe quel caractère zéro ou plusieurs fois
txt = '''Apple and banana are fruits'''
matches = re.findall(regex_pattern, txt)
print(matches) 
```

### Zero or one time(?)
Zero or one time. The pattern may not occur or it may occur once.

### <font color="blue">Zéro ou une fois (?)
<font color="blue">Zéro ou une fois. Le schéma peut ne pas se produire ou se produire une seule fois.

```python
txt = '''I am not sure if there is a convention how to write the word e-mail.
Some people write it as email others may write it as Email or E-mail.'''
# ? means here that '-' is optional / ? signifie ici que '-' est facultatif
regex_pattern = r'[Ee]mail'  
matches = re.findall(regex_pattern, txt)
print(matches)
```

```python
txt = '''I am not sure if there is a convention how to write the word e-mail.
Some people write it as email others may write it as Email or E-mail.'''
# ? means here that '-' is optional / ? signifie ici que '-' est facultatif
regex_pattern = r'[Ee]-mail'  
matches = re.findall(regex_pattern, txt)
print(matches)
```

```python
txt = '''I am not sure if there is a convention how to write the word e-mail.
Some people write it as email others may write it as Email or E-mail.'''
# ? means here that '-' is optional / ? signifie ici que '-' est facultatif
regex_pattern = r'[Ee]-?mail'  
matches = re.findall(regex_pattern, txt)
print(matches)
```

### Quantifier in RegEx
We can specify the length of the substring we are looking for in a text, using a curly bracket. Let us imagine, we are interested in a substring with a length of 4 characters:

### <font color="blue"> Quantificateur dans RegEx
<font color="blue">Nous pouvons spécifier la longueur de la sous-chaîne recherchée dans un texte à l'aide d'accolades. Imaginons que nous soyons intéressés par une sous-chaîne de 4 caractères :


```python
txt = 'This regular expression example was made on December 6,  2019 and revised on July 8, 2021'
regex_pattern = r'\d{4}'  # exactly four times / exactement quatre fois
matches = re.findall(regex_pattern, txt)
print(matches) 
```

```python

txt = 'This regular expression example was made on December 6, 2019 and revised on July 8, 2021, 234'
regex_pattern = r'\d{1,3}'   
matches = re.findall(regex_pattern, txt)
print(matches) 
```

### Cart ^

```python
txt = 'This regular expression example was made on December 6,  2019 and revised on July 8, 2021'
regex_pattern = r'^This'  # ^ means starts with / ^ signifie commence par
matches = re.findall(regex_pattern, txt)
print(matches)  
```

### Negation

```python
txt = 'This regular expression example was made on December 6,  2019 and revised on July 8, 2021'
regex_pattern = r'[^A-Za-z ]+'  
# ^ in set character means negation, not A to Z, not a to z, no space
# ^ dans le caractère set signifie négation, pas de A à Z, pas de a à z, pas d'espace
matches = re.findall(regex_pattern, txt)
print(matches)
```
