### Mathematical Operation using numpy
NumPy array is not like exactly like python list. To do mathematical operation in Python list we have to loop through the items but numpy can allow to do any mathematical operation without looping. Mathematical Operation:
<br>Addition (+)
<br>Subtraction (-)
<br>Multiplication (*)
<br>Division (/)
<br>Modules (%)
<br>Floor Division(//)
<br>Exponential(**)

### <font color="blue"> Opérations mathématiques avec Numpy
<font color="blue">Un tableau Numpy est différent d'une liste Python. Pour effectuer des opérations mathématiques dans une liste Python, il faut parcourir les éléments, mais Numpy permet d'effectuer n'importe quelle opération mathématique sans boucle. Opérations mathématiques :
<br><font color="blue">Addition (+)
<br><font color="blue">Soustraction (-)
<br><font color="blue">Multiplication (*)
<br><font color="blue">Division (/)
<br><font color="blue">Modules (%)
<br><font color="blue">Division par le bas (//)
<br><font color="blue">Exponentielle (**)

```python
# Addition
numpy_array_from_list = np.array([1, 2, 3, 4, 5])
print('original array: ', numpy_array_from_list)
```

```python
ten_plus_original = numpy_array_from_list  + 10
print(ten_plus_original)
```

```python
# Subtraction
numpy_array_from_list = np.array([1, 2, 3, 4, 5])
print('original array: ', numpy_array_from_list)
```

```python
ten_minus_original = numpy_array_from_list  - 10
print(ten_minus_original)
```

```python
# Multiplication
numpy_array_from_list = np.array([1, 2, 3, 4, 5])
print('original array: ', numpy_array_from_list)
```

```python
ten_times_original = numpy_array_from_list * 10
print(ten_times_original)
```

```python
# Division
numpy_array_from_list = np.array([1, 2, 3, 4, 5])
print('original array: ', numpy_array_from_list)
```

```python
ten_times_original = numpy_array_from_list / 10
print(ten_times_original)
```

```python
# Modulus; Finding the remainder
# Module ; Trouver le reste
numpy_array_from_list = np.array([1, 2, 3, 4, 5])
print('original array: ', numpy_array_from_list)
```

```python
ten_times_original = numpy_array_from_list % 3
print(ten_times_original)
```

```python
# Floor division: the division result without the remainder
# Division au sol : le résultat de la division sans le reste
numpy_array_from_list = np.array([1, 2, 3, 4, 5])
print('original array: ', numpy_array_from_list)
```

```python
ten_times_original = numpy_array_from_list // 10
print(ten_times_original)
```

```python
# Exponential is finding some number the power of another:
# L'exponentielle consiste à trouver à un nombre la puissance d'un autre :
numpy_array_from_list = np.array([1, 2, 3, 4, 5])
print('original array: ', numpy_array_from_list)
```

```python
ten_times_original = numpy_array_from_list  ** 2
print(ten_times_original)
```
