### While Loop
We use the reserved word while to make a while loop. It is used to execute a block of statements repeatedly until a given condition is satisfied. When the condition becomes false, the lines of code after the loop will be continued to be executed.

============<br>

### <font color="blue">Boucle While
<font color="blue">Nous utilisons le mot réservé while pour créer une boucle while. Il permet d'exécuter un bloc d'instructions de manière répétée jusqu'à ce qu'une condition donnée soit satisfaite. Lorsque la condition devient fausse, l'exécution des lignes de code suivant la boucle se poursuit.

```python
count = 0
while count <= 7:
    print(count)
    count = count + 1
```

```python
# The below loop condition will be false when count is 5 and the loop stops, and execution starts the else statement. As a result 5 will be printed.
# La condition de boucle ci-dessous sera fausse si count vaut 5, la boucle s'arrêtera et l'exécution démarrera l'instruction else. Le résultat sera 5.
count = 0
while count < 5:
    print(count)
    count = count + 1
else:
    print(count)
```

### Break
Break: We use break when we like to get out of or stop the loop.
    
=========<br>
### <font color="blue">Pause
<font color="blue">Pause : On utilise pause lorsque l’on souhaite sortir ou arrêter la boucle.

```python
# syntax
while condition:
    code goes here
    if another_condition:
        break
```

```python
# The below while loop only prints 0, 1, 2, but when it reaches 3 it stops.
# La boucle while ci-dessous imprime uniquement 0, 1, 2, mais lorsqu'elle atteint 3, elle s'arrête.
count = 0
while count < 5:
    print(count)
    count = count + 1
    if count == 3:
        break
```

### Continue: 
With the continue statement we can skip the current iteration, and continue with the next

========<br>
### <font color="blue">Continuer :
<font color="blue">Avec l'instruction continue, nous pouvons ignorer l'itération en cours et passer à la suivante.

```python
 # syntax
while condition:
    code goes here
    if another_condition:
        continue
```

```python
# The below while loop only prints 0, 1, 2 and 4 (skips 3).
# La boucle while ci-dessous imprime uniquement 0, 1, 2 et 4 (saute 3).
count = 0
while count < 5:
    if count ==2:
        count = count + 1
        continue
    print(count)
    count = count + 1
    
```

### for Loop

```python
# syntax
for iterator in lst:
    code goes here
```

```python
# Iterating over a list / Itérer sur une liste

numbers = [1, 2, 3, 4, 5] 
# number is temporary name to refer to the list's items, valid only inside this loop
# number est un nom temporaire pour faire référence aux éléments de la liste, valide uniquement à l'intérieur de cette boucle
for y in numbers:
    print(y)
```

# The Range Function

Par défaut, elle commence à 0 et l'incrément est de 1. La séquence range nécessite au moins un argument (end). Création de séquences avec range
    
    

```python
#range(start, end, step)

lst = list(range(11)) 
print(lst, '\n') 
print('==============','\n') 
# 2 arguments indicate start and end of the sequence, step set to default 1
# 2 arguments indiquent le début et la fin de la séquence, l'étape est définie par défaut sur 1

st = set(range(1, 11))    
print(st, '\n') 

print('==============','\n') 

lst = list(range(0,11,3))
print("Ma liste avec 3 comme pas: ",lst, '\n') 

st = set(range(0,11,2))
print(st) 
```

### Nested Loop
In a nested loop, the first loop is called the outer loop, and the loop inside is the inner loop. So, for every iteration of the outer loop, the inner loop runs completely before the outer loop moves to the next iteration.

==============
### <font color="blue">Boucle imbriquée
<font color="blue">Dans une boucle imbriquée, la première boucle est appelée boucle externe, et la boucle interne est appelée boucle interne. Ainsi, à chaque itération de la boucle externe, celle-ci s'exécute complètement avant de passer à l'itération suivante.

```python
for outer_variable in outer_iterable:
    for inner_variable in inner_iterable:
        print("Body goes here")
```

```python
#Here’s how the clock logic looks in Python code:/ 
# Voici à quoi ressemble la logique de l’horloge dans le code Python :
for hour in range(0, 24):
    for minute in range(0, 60):
        print(f"{hour:02d}:{minute:02d}")
```

Another application of nested loops is processing elements in more than one dimension. A multidimensional or nested list is a list that contains other lists as its elements. To sum all elements in a nested list, you can use nested loops.

=========

<font color="blue">Une autre application des boucles imbriquées est le traitement d'éléments multidimensionnels. Une liste multidimensionnelle, ou imbriquée, est une liste dont les éléments sont d'autres listes. Pour additionner tous les éléments d'une liste imbriquée, vous pouvez utiliser des boucles imbriquées.

```python
# Summing Elements in Multidimensional Lists / Sommation d'éléments dans des listes multidimensionnelles
resource_donators = [
     [8, 6, 3], 
     [9, 2, 7],
     [4, 1, 5]
    ]

total_resources = 0 # initialize the total resources value first / initialiser d'abord la valeur totale des ressources
for planet in resource_donators:
     for resource in planet:
            total_resources += resource
print(f"All resources gathered for interstellar travels: {total_resources}")
```

### Replacing Nested Loops With List Comprehension
With list comprehensions, you can replace the intimidating hill of nested loops with a more compact and readable expression. While list comprehensions don’t necessarily improve performance or reduce time complexity—especially in nested iterations—they often make your intent clearer and your code easier to understand.

===========<br>
### <font color="blue">Replacing Nested Loops With List Comprehension
<font color="blue">Grâce aux compréhensions de liste, vous pouvez remplacer l'impressionnante montagne de boucles imbriquées par une expression plus compacte et lisible. Si les compréhensions de liste n'améliorent pas nécessairement les performances ni ne réduisent la complexité temporelle, surtout dans les itérations imbriquées, elles clarifient souvent votre intention et facilitent la compréhension de votre code.

```python
bases = ["plain", "chocolate", "blueberry"]
toppings = ["honey", "whipped cream", "alien syrup"]
pancake_stacks = []

for base in bases:
     for topping in toppings:
        pancake_stacks.append(f"{base.capitalize()} pancake with {topping}")

print(pancake_stacks)
```

```python
# The above could be replaced by the below script: / Le script ci-dessus pourrait être remplacé par le script ci-dessous :
bases = ["plain", "chocolate", "blueberry"]
toppings = ["honey", "whipped cream", "alien syrup"]
pancake_stacks = [
     f"{base.capitalize()} pancake with {topping}"
     for base in bases
     for topping in toppings
 ]
print(pancake_stacks)
```

### Combining function and Iterations 

### <font color="blue">Combinaison de fonctions et d'itérations

```python
# writing function count_to_n and using it 
def count_to_n(n):
    for i in range(1, n + 1):
        print(i, end=' ')
    print()
    
for i in range(1, 10):
    count_to_n(i)
```

```python
# Get tree height from user
height = int(input('Enter height of tree: '))
# Draw one row for every unit of height
row = 0
while row < height:
# Print leading spaces; as row gets bigger, the number of
# leading spaces gets smaller
    count = 0
    while count < height - row:
        print(end=' ')
        count += 1
# Print out stars, twice the current row plus one:
# 1. number of stars on left side of tree
# = current row value
# 2. exactly one star in the center of tree
# 3. number of stars on right side of tree
# = current row value
    count = 0
    while count < 2*row + 1:
        print(end='*')
        count += 1
# Move cursor down to next line
    print()
    row += 1 # Consider next row
```

```python
# Bringing everything together / Rassembler tout ensemble
def process_grades(students):
    """Function to process students' grades and determine if they passed or failed."""
    """Fonction permettant de traiter les notes des étudiants et de déterminer leur réussite ou leur échec.""" 
    passing_grade = 50  # Define the passing grade / Définir la note de passage

    for student, grade in students:
        if grade >= passing_grade:
            result = "Passed"
        else:
            result = "Failed"
        print(f"Student: {student}, Grade: {grade}, Result: {result}")

# List of students and their grades / Liste des étudiants et leurs notes
students_list = [("Alice", 85), ("Bob", 42), ("Charlie", 78), ("Diana", 65), ("Eve", 30)]

# Call the function to process the grades / Appeler la fonction pour traiter les notes
process_grades(students_list)
```
