Python uses try and except to handle errors gracefully. A graceful exit (or graceful handling) of errors is a simple programming idiom - a program detects a serious error condition and "exits gracefully", in a controlled manner as a result. Often the program prints a descriptive error message to a terminal or log as part of the graceful exit, this makes our application more robust. The cause of an exception is often external to the program itself. An example of exceptions could be an incorrect input, wrong file name, unable to find a file, a malfunctioning IO device. Graceful handling of errors prevents our applications from crashing.

We have covered the different Python error types in the previous section. If we use try and except in our program, then it will not raise errors in those blocks.

```python
try:
    name = input('Enter your name:')
    year_born = input('Year you born:')
    age = 2019 - int(year_born)
    print(f'You are {name}. And your age is {age}.')
except TypeError:
    print('Type error occur')
except ValueError:
    print('Value error occur')
except ZeroDivisionError:
    print('zero division error occur')
else:
    print('I usually run with the try block')
finally:
    print('I alway run.')
```

### Packing and Unpacking Arguments in Python
We use two operators:
<ul>
<li>* for tuples</li>
<li>** for dictionaries</li></ul>
Let us take as an example below. It takes only arguments but we have list. We can unpack the list and changes to argument.

### <font color="blue">Empaquetage et dépaquetage des arguments en Python</font>
<font color="blue">Nous utilisons deux opérateurs :</font>
<ul><font color="blue">
<li>* pour les tuples</li>
<li>** pour les dictionnaires</li> </font></ul>
<font color="blue">Prenons l'exemple ci-dessous. Il ne prend que des arguments, mais nous avons une liste. Nous pouvons dépaqueter la liste et modifier l'argument.</font>

```python
# unpacking a list / depackez une liste
def sum_of_five_nums(a, b, c, d, e):
    return a + b + c + d + e

lst = [1, 2, 3, 4, 5]
print(sum_of_five_nums(lst))
```

```python
# When we run the this code, it raises an error, because this function takes numbers (not a list) as arguments. Let us unpack/destructure the list.
# L'exécution de ce code génère une erreur, car cette fonction prend des nombres (et non une liste) comme arguments. Décompressons/déstructurons la liste.
def sum_of_five_nums(a, b, c, d, e):
    return a + b + c + d + e

lst = [1, 2, 3, 4, 5]
print(sum_of_five_nums(*lst))
```

```python
# normal call with separate arguments / appel normal avec des arguments séparés
numbers = range(2, 7)  
print(list(numbers))

```

```python
# call with arguments unpacked from a list / appel avec des arguments décompressés d'une liste
args = [2, 7]
numbers = range(*args)  
print(numbers)      
```

```python
# A list or a tuple can also be unpacked like this:
# Une liste ou un tuple peut également être décompressé comme ceci :
countries = ['New Jersey', 'New York', 'Connecticut', 'Denmark', 'Iceland', 'Ghana', 'Togo']
nj, ny, cn, *rest = countries
print(nj, ny, cn, rest)   
```

```python

numbers = [1, 2, 3, 4, 5, 6, 7]
one, *middle, last = numbers
print(one, middle, last)  
```

```python
# Unpacking Dictionaries / Déballage des dictionnaires
def unpacking_person_info(name, country, city, age):
    return f'{name} lives in {country}, {city}. He is {age} year old.'
```

```python
dct = {'name':'Chayil', 'country':'USA', 'city':'NYC', 'age':250}
print(unpacking_person_info(**dct))
```

### Packing
Sometimes we never know how many arguments need to be passed to a python function. We can use the packing method to allow our function to take unlimited number or arbitrary number of arguments.

### <font color="blue">Empaquetage</font>
<font color="blue">Parfois, nous ne savons jamais combien d'arguments doivent être passés à une fonction Python. Nous pouvons utiliser la méthode d'empilage pour permettre à notre fonction de prendre un nombre illimité ou arbitraire d'arguments.

```python
def sum_all(*args):
    s = 0
    for i in args:
        s += i
    return s
print(sum_all(1, 2, 3))             
print(sum_all(1, 2, 3, 4, 5, 6, 7))
```

```python
def packing_person_info(**kwargs):
    # check the type of kwargs and it is a dict type
    # print(type(kwargs))
    # Printing dictionary items
    for key in kwargs:
        print(f"{key} = {kwargs[key]}")
    return kwargs

print(packing_person_info(name="Chayil",
      country="USA", city="NYC", age=250))
```
