### Function to display weekly dates for a given year

### Fonction permettant d'afficher les dates hebdomadaires pour une ann√©e donn√©e

```python
from datetime import datetime, timedelta

def display_weekly_dates(year):
    """
    Display all Monday dates for a given year in a clean format.
    Uses Monday as the start of each week.
    """
    # Start with the first Monday of the year
    jan_1 = datetime(year, 1, 1)
    days_to_monday = (7 - jan_1.weekday()) % 7
    first_monday = jan_1 + timedelta(days=days_to_monday)
    
    # If January 1st is already a Monday, use it
    if jan_1.weekday() == 0:
        first_monday = jan_1
    
    print(f"üìÖ Weekly Dates for {year}")
    print("=" * 40)
    
    current_date = first_monday
    week_number = 1
    
    while current_date.year == year:
        # Format: Week 1: Jan 02 - Jan 08
        week_end = current_date + timedelta(days=6)
        
        # Handle year transition for the last week
        if week_end.year != year:
            week_end = datetime(year, 12, 31)
        
        print(f"Week {week_number:2d}: {current_date.strftime('%b %d')} - {week_end.strftime('%b %d')}")
        
        # Move to next Monday
        current_date += timedelta(weeks=1)
        week_number += 1
    
    print(f"\n‚ú® Total weeks in {year}: {week_number - 1}")

# Example usage
if __name__ == "__main__":
    # Display weekly dates for current year
    current_year = datetime.now().year
    display_weekly_dates(current_year)
    
    # You can also try other years
    print("\n" + "="*50 + "\n")
    display_weekly_dates(2024)
```

### Weekly date for weekdays

### Date hebdomadaire pour les jours de la semaine

Display all 5-day workweeks (Monday-Friday) for a given year.</br>
<p>Shows each workweek as a clean date range.</p><br>

<font color="blue">
Affiche toutes les semaines de travail de 5 jours (du lundi au vendredi) pour une ann√©e donn√©e.

Affiche chaque semaine de travail sous forme de plage de dates propre.
</font>


```python
from datetime import datetime, timedelta

def display_workweek_dates(year):
    
    # Start with the first Monday of the year
    # Commencez par le premier lundi de l'ann√©e
    jan_1 = datetime(year, 1, 1)
    days_to_monday = (7 - jan_1.weekday()) % 7
    first_monday = jan_1 + timedelta(days=days_to_monday)
    
    # If January 1st is already a Monday, use it
    # Si le 1er janvier est d√©j√† un lundi, utilisez-le
    if jan_1.weekday() == 0:
        first_monday = jan_1
    
    print(f"üíº Workweek Calendar for {year}")
    print("=" * 45)
    print("üìç Monday ‚Üí Friday (5-day workweeks)")
    print("-" * 45)
    
    current_monday = first_monday
    week_number = 1
    
    while current_monday.year == year:
        # Calculate Friday (4 days after Monday)
        # Calculer vendredi (4 jours apr√®s lundi)
        friday = current_monday + timedelta(days=4)
        
        # Handle year transition
        # G√©rer la transition d'ann√©e
        if friday.year != year:
            # Find the last workday of the year
            # Trouver le dernier jour ouvrable de l'ann√©e
            last_day = datetime(year, 12, 31)
            # If last day is weekend, find previous Friday
            # Si le dernier jour est un week-end, recherchez le vendredi pr√©c√©dent
            while last_day.weekday() > 4:  # 5=Saturday, 6=Sunday
                last_day -= timedelta(days=1)
            friday = last_day
        
        # Format the workweek
        # Formater la semaine de travail
        print(f"Week {week_number:2d}: {current_monday.strftime('%b %d')} ‚Üí {friday.strftime('%b %d')} "
              f"({current_monday.strftime('%a')} - {friday.strftime('%a')})")
        
        # Move to next Monday
        # D√©placer √† lundi prochain
        current_monday += timedelta(weeks=1)
        week_number += 1
        
        # Stop if we've gone beyond the year
        # Arr√™tez si nous avons d√©pass√© l'ann√©e
        if current_monday.year > year:
            break
    
    # Calculate total workdays
    # Calculer le nombre total de jours de travail
    total_workdays = (week_number - 1) * 5
    print(f"\nüìä Summary:")
    print(f"   ‚Ä¢ Total workweeks: {week_number - 1}")
    print(f"   ‚Ä¢ Total workdays: ~{total_workdays} days")
    print(f"   ‚Ä¢ Weekend days: ~{52 * 2} days")

def display_workdays_by_month(year, month):
    """
    Bonus: Display workdays for a specific month
    """
    month_start = datetime(year, month, 1)
    # Find the last day of the month
    # Trouver le dernier jour du mois
    if month == 12:
        month_end = datetime(year + 1, 1, 1) - timedelta(days=1)
    else:
        month_end = datetime(year, month + 1, 1) - timedelta(days=1)
    
    print(f"\nüí° Workdays in {month_start.strftime('%B %Y')}:")
    print("-" * 30)
    
    current_date = month_start
    workdays = []
    
    while current_date <= month_end:
        if current_date.weekday() < 5:  # Monday=0, Friday=4
            workdays.append(current_date.strftime('%a %d'))
        current_date += timedelta(days=1)
    
    # Display in rows of 5 (one workweek per row)
    # Affichage en lignes de 5 (une semaine de travail par ligne)
    for i in range(0, len(workdays), 5):
        week_days = workdays[i:i+5]
        print(f"  {' | '.join(f'{day:>6}' for day in week_days)}")
    
    print(f"\n  Total workdays: {len(workdays)}")

# Example usage
# Exemple d'utilisation
if __name__ == "__main__":
    # Display workweeks for current year
    # Afficher les semaines de travail pour l'ann√©e en cours
    current_year = datetime.now().year
    display_workweek_dates(current_year)
    
    # Bonus: Show current month's workdays
    # Bonus¬†: Afficher les jours ouvrables du mois en cours
    current_month = datetime.now().month
    display_workdays_by_month(current_year, current_month)
```
